```python
# CLI To-Do App With File Persistence
# todo.py

import json
import os

# ── NODE ──────────────────────────────────────────────
class Node:
    def __init__(self, task):
        self.task = task        # the actual to-do text
        self.next = None        # pointer to next node

# ── LINKED LIST ───────────────────────────────────────
class LinkedList:
    def __init__(self):
        self.head = None        # start of the chain

    def add(self, task):
        new_node = Node(task)
        if not self.head:
            self.head = new_node
            return
        current = self.head
        while current.next:     # walk to the end
            current = current.next
        current.next = new_node # attach new node

    def delete(self, task):
        if not self.head:
            return False
        # special case — deleting the first node
        if self.head.task == task:
            self.head = self.head.next
            return True
        current = self.head
        while current.next:
            if current.next.task == task:
                current.next = current.next.next  # skip over it
                return True
            current = current.next
        return False

    def display(self):
        tasks = []
        current = self.head
        while current:
            tasks.append(current.task)
            current = current.next
        return tasks

    def to_list(self):          # for saving to file
        return self.display()

    def from_list(self, items): # for loading from file
        for item in items:
            self.add(item)

# ── STACK (for undo/redo) ─────────────────────────────
class Stack:
    def __init__(self):
        self.items = []

    def push(self, action):
        self.items.append(action)

    def pop(self):
        if self.is_empty():
            return None
        return self.items.pop()

    def is_empty(self):
        return len(self.items) == 0

# ── FILE PERSISTENCE ──────────────────────────────────
FILE = "todos.json"

def save(todo_list):
    with open(FILE, "w") as f:
        json.dump(todo_list.to_list(), f)

def load(todo_list):
    if os.path.exists(FILE):
        with open(FILE, "r") as f:
            items = json.load(f)
            todo_list.from_list(items)

# ── MAIN APP ──────────────────────────────────────────
def main():
    todo_list = LinkedList()
    undo_stack = Stack()
    redo_stack = Stack()

    load(todo_list)             # load saved tasks on startup
    print("\n📝 To-Do App | commands: add / delete / list / undo / redo / quit\n")

    while True:
        command = input("→ ").strip().lower()

        if command == "add":
            task = input("  Task: ").strip()
            todo_list.add(task)
            undo_stack.push(("add", task))  # remember this action
            redo_stack = Stack()            # clear redo on new action
            save(todo_list)
            print(f"  ✅ Added: {task}")

        elif command == "delete":
            task = input("  Task to delete: ").strip()
            if todo_list.delete(task):
                undo_stack.push(("delete", task))
                redo_stack = Stack()
                save(todo_list)
                print(f"  🗑️  Deleted: {task}")
            else:
                print("  ❌ Task not found")

        elif command == "list":
            tasks = todo_list.display()
            if tasks:
                print("\n  Your tasks:")
                for i, t in enumerate(tasks, 1):
                    print(f"  {i}. {t}")
                print()
            else:
                print("  No tasks yet\n")

        elif command == "undo":
            action = undo_stack.pop()
            if action:
                kind, task = action
                if kind == "add":
                    todo_list.delete(task)  # undo an add = delete
                elif kind == "delete":
                    todo_list.add(task)     # undo a delete = add back
                redo_stack.push(action)
                save(todo_list)
                print(f"  ↩️  Undid: {kind} '{task}'")
            else:
                print("  Nothing to undo")

        elif command == "redo":
            action = redo_stack.pop()
            if action:
                kind, task = action
                if kind == "add":
                    todo_list.add(task)     # redo an add = add again
                elif kind == "delete":
                    todo_list.delete(task)  # redo a delete = delete again
                undo_stack.push(action)
                save(todo_list)
                print(f"  ↪️  Redid: {kind} '{task}'")
            else:
                print("  Nothing to redo")

        elif command == "quit":
            print("  Bye! 👋")
            break

        else:
            print("  Unknown command")

if __name__ == "__main__":
    main()
```
